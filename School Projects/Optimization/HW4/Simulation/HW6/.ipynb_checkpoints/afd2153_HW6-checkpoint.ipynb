# Problem 1

```python
def f(x):
    return np.exp(np.exp(x))

def monte_carlo_integration(N):
    sum_f = 0
    f_values = []
    
    for i in range(N):
        X = np.random.uniform(0, 1)
        fx = f(X)
        f_values.append(fx)
        sum_f += fx
    
    estimate = sum_f / N
    
    std_dev = np.std(f_values)
    standard_error = std_dev / np.sqrt(N)
    
    confidence_interval = (estimate - 1.96 * standard_error, estimate + 1.96 * standard_error)    
    
    return (estimate, confidence_interval)

N = 1000

estimate, confidence_interval = monte_carlo_integration(N)

actual_integral, _ = quad(f, 0, 1)

print(f"Monte Carlo Estimated Integral: {estimate}")
print(f"Actual value of the Integral (from scipy): {actual_integral}")
print(f"95% Confidence Interval: {confidence_interval}")
```

```python
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Compute the actual integral using scipy's quad function
actual_integral, _ = quad(f, 0, 1)

sample_sizes = np.arange(100, 10001, 1)

errors = []

for N in sample_sizes:
    monte_carlo_estimate = monte_carlo_integration(N)[0]
    # Calculate the percentage error
    error = abs((monte_carlo_estimate - actual_integral) / actual_integral) * 100
    errors.append(error)

plt.figure(figsize=(8, 5))
plt.plot(sample_sizes, errors, label="Percentage Error", color='blue')
plt.title("Percentage Error of Monte Carlo Estimate vs. Number of Samples")
plt.xlabel("Number of Samples")
plt.ylabel("Percentage Error (%)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def f(x):
    return np.exp(x + x**2)

def monte_carlo_integration_b(N):
    sum_f = 0
    f_values = []
    
    for i in range(N):
        U = np.random.uniform(0, 1)
        X = 4 * U - 2
        fx = f(X)
        f_values.append(fx)
        sum_f += fx
    
    estimate = 4 * sum_f / N
    
    std_dev = np.std(f_values)
    standard_error = std_dev / np.sqrt(N)
    
    confidence_interval = (estimate - 1.96 * standard_error, estimate + 1.96 * standard_error)    
    return estimate, confidence_interval

N = 1000

estimate_b, confidence_interval_b = monte_carlo_integration_b(N)

actual_integral_b, _ = quad(f, -2, 2)

print(f"Monte Carlo Estimated Integral: {estimate_b}")
print(f"Actual value of the Integral (from scipy): {actual_integral_b}")
print(f"95% Confidence Interval: {confidence_interval_b}")
```

```python
import numpy as np
from scipy.integrate import quad

def f(x):
    return 1 / x2

def transformed_function(U):
    x = U / (1 - U)
    return f(x) / (1 - U)**2

def monte_carlo_integration_c(N):
    sum_f = 0
    f_values = []
    
    for i in range(N):
        U = np.random.uniform(0, 1)
        fx = transformed_function(U)
        f_values.append(fx)
        sum_f += fx
    
    estimate = sum_f / N
    
    std_dev = np.std(f_values)
    standard_error = std_dev / np.sqrt(N)
    
    confidence_interval = (estimate - 1.96 * standard_error, estimate + 1.96 * standard_error)
    
    return (estimate, confidence_interval)

N = 1000

estimate_c, confidence_interval_c = monte_carlo_integration_c(N)

actual_integral_c, _ = quad(f, 0, np.inf)

print(f"Monte Carlo Estimated Integral: {estimate_c}")
print(f"95% Confidence Interval: {confidence_interval_c}")
```

# Problem 2

```python
import random
import math

num_simulations = 1000

N = 1000

cov_U_eU_list = []
correlation_list = []

def calculate_cov_corr(U_values):
    eU_values = [math.exp(U) for U in U_values]
    sqrt_1_minus_U2_values = [math.sqrt(1 - U**2) for U in U_values]
    
    mean_U = sum(U_values) / N
    mean_eU = sum(eU_values) / N
    mean_sqrt = sum(sqrt_1_minus_U2_values) / N

    cov_U_eU = sum((U_values[i] - mean_U) * (eU_values[i] - mean_eU) for i in range(N)) / N

    cov_U_sqrt = sum((U_values[i] - mean_U) * (sqrt_1_minus_U2_values[i] - mean_sqrt) for i in range(N)) / N

    std_U = math.sqrt(sum((U - mean_U)**2 for U in U_values) / N)
    std_sqrt = math.sqrt(sum((v - mean_sqrt)**2 for v in sqrt_1_minus_U2_values) / N)

    correlation = cov_U_sqrt / (std_U * std_sqrt)

    return cov_U_eU, correlation

for _ in range(num_simulations):
    U_values = [random.uniform(0, 1) for _ in range(N)]
    
    cov_U_eU, correlation = calculate_cov_corr(U_values)
    
    cov_U_eU_list.append(cov_U_eU)
    correlation_list.append(correlation)

point_estimate_cov_U_eU = sum(cov_U_eU_list) / num_simulations
point_estimate_correlation = sum(correlation_list) / num_simulations

print(f"Point Estimate of Cov(U, e^U): {point_estimate_cov_U_eU}")
print(f"Point Estimate of Corr(U, sqrt(1 - U^2)): {point_estimate_correlation}")
```
